```python
def nullvaluesfinder(df):
    nullcolumns=df.columns[df.isnull().any()].tolist()
    nullcoldf=pd.DataFrame(df[nullcolumns].isnull().sum()/len(df)*100).reset_index()
    nullcoldf.columns=['colname','missingpercentage']
    return nullcoldf
    
```

```python
nullcoldf=nullvaluesfinder(traindf)
#nullcolumns=traindf.columns[traindf.isnull().any()].tolist()
#nullcoldf=pd.DataFrame(traindf[nullcolumns].isnull().sum()/len(traindf)*100).reset_index()
#nullcoldf.columns=['colname','missingpercentage']
```
